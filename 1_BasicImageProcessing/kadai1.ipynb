```python
img = cv2.imread('imori_256x256.png') # BGR
img = img[..., ::-1] # BGR -> RGB

plt.figure(figsize=(7,7))
plt.imshow(img)
print(img.shape)
```

```python
resized_size = (128,128)
resized_img = cv2.resize(img, dsize=resized_size, interpolation=cv2.INTER_LINEAR) # resized sizeにリサイズ

plt.figure(figsize=(7,7))
plt.imshow(resized_img)
print(resized_img.shape)
```

```python
rotation_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) #時計回りに90度回転

plt.figure(figsize=(7,7))
plt.imshow(rotation_img)
print(rotation_img.shape)
```
