```python
img = cv2.imread('imori_256x256.png') # BGR
img = img[..., ::-1] # BGR -> RGB

plt.imshow(img)
print(img.shape)
```

```python
resized_size = (128,128)
resized_img = cv2.resize(img, dsize=resized_size, interpolation=cv2.INTER_LINEAR) # resized size

plt.imshow(resized_img)
print(resized_img.shape)
```

```python
rotation_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) #90

plt.imshow(rotation_img)
print(rotation_img.shape)
```

```python
gray_img = cv2.imread("imori_256x256.png", 0)
ret2, binary_img = cv2.threshold(gray_img, 0, 255, cv2.THRESH_OTSU) #

plt.imshow(binary_img, cmap="gray")
print(binary_img.shape)
```

```python
lena = cv2.imread("len_std.jpg")[..., ::-1]
imori = cv2.imread("imori_256x256.png")[..., ::-1]

if lena.shape == imori.shape:
    imaga_sa = np.absolute(lena-imori)
    plt.imshow(imaga_sa)

else:
    print("Image sizes are different")
```
